```python
# prompt: 2개의 정수를 입력받고 최대공배수를 구한 뒤 하나의 그룹에 결과를 저장하고, 다시 2개의 정수를 입력받고 최대공배수의 결과를 다른 그룹에 저장하는 방식을 총 4번 반복한 후 모든 그룹의 최대공배수를 비교해서 수가 큰 순서대로 정렬하고, 각자 입력받은 정수와 그룹 이름을 차트에서 한번에 확인 할 수 있도록 비교하는 차트를 만들어줘

import math

def gcd(a, b):
  if b == 0:
    return a
  return gcd(b, a % b)

def lcm(a, b):
  return (a * b) // gcd(a, b)

groups = []

for i in range(4):
  while True:
    try:
      num1 = int(input(f"Group {i+1}: Enter the first integer: "))
      num2 = int(input(f"Group {i+1}: Enter the second integer: "))
      break
    except ValueError:
      print("Invalid input. Please enter integers only.")

  result = lcm(num1, num2)
  groups.append({"group": i+1, "num1": num1, "num2": num2, "lcm": result})


sorted_groups = sorted(groups, key=lambda x: x["lcm"], reverse=True)

# Visualization using matplotlib
import matplotlib.pyplot as plt

group_names = [f"Group {group['group']}" for group in sorted_groups]
lcm_values = [group["lcm"] for group in sorted_groups]
num1_values = [group["num1"] for group in sorted_groups]
num2_values = [group["num2"] for group in sorted_groups]

plt.figure(figsize=(10, 6))
plt.bar(group_names, lcm_values, color='skyblue')

# Add values on top of each bar
for i, v in enumerate(lcm_values):
  plt.text(i, v + 1, str(v), ha='center', va='bottom')

# Add num1 and num2 values as labels below the bars
for i in range(len(group_names)):
  plt.text(i, 0, f"({num1_values[i]}, {num2_values[i]})", ha='center', va='top')


plt.xlabel("Groups")
plt.ylabel("Least Common Multiple (LCM)")
plt.title("Comparison of LCMs across Groups")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
